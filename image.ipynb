```python
import cv2
import numpy as np


drawing = False
ix = -1
iy = -1


def draw_rectangle(event,x,y,flags,params):
    global ix,iy,drawing
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy),(x,y), (0,255,0),-1)
    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)

            
img = np.zeros((1512,1512,3))

cv2.namedWindow(winname='drawing2')
cv2.setMouseCallback('drawing2', draw_rectangle)


while True:
    cv2.imshow('drawing2', img)

    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

#### Image Basics


```python
import cv2
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
my_image = np.zeros(shape =(512,512,3),dtype=np.int16)
my_image.shape
```

```python
plt.imshow(my_image)
```

```python
cv2.rectangle(my_image, pt1 = (10,10), pt2 = (502,502), color = (255,120,0), thickness=5)
plt.imshow(my_image)
```

```python
cv2.circle(my_image, center = (250,250) ,radius=70, color = (0,255,80), thickness = -1)
plt.imshow(my_image)
```

```python
cv2.circle(my_image, center = (250,250), radius=90,color = (0,255,80), thickness = 5)
cv2.circle(my_image, center = (250,250), radius=110,color = (80,80,255), thickness = 5)

plt.imshow(my_image)
```

```python
plt.imshow(my_image)
```

```python
cv2.line(my_image, pt1 = (10,10), pt2 = (170,170), color=(250,250,120), thickness=5)
cv2.line(my_image, pt1 = (330,330), pt2 = (502,502), color=(250,250,120), thickness=5)
cv2.line(my_image, pt1 = (10,502), pt2= (170,332), color=(250,250,120), thickness=5)
cv2.line(my_image, pt1 = (502,10), pt2= (332,170), color=(250,250,120), thickness=5)



plt.imshow(my_image)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(my_image, text = 'Art Here', org=(120,480),fontFace=font,fontScale=2,color=(255,255,255),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(my_image)
```
