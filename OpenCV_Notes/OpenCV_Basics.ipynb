## Drawing Images


```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img, pt1=(100,100),pt2=(250,250), color=(0,255,0),thickness=8)
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(300,400), radius=80, color=(10,150,200),thickness=3)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img, center=(190,190), radius=20, color=(100,100,100),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(img=blank_img,pt1=(100,100),pt2=(250,250),color=(120,30,30),thickness=6)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text="Computer Vision",org=(160,50), fontFace=font,fontScale=1,color=(100,100,200),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
vertices
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,100,100),thickness=5)
plt.imshow(blank_img)
```
